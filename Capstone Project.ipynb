```python
#I assume that there is a postively correlated relationship between a weapon's bullet
#speed and it's range. Because if you would like to kill the enemy at a long distance,
#the bullet speed must be fast enough.
#Now I would like to explore that is there any relationship between a weapon's bullet 
#speed and it's range.
X = df['Bullet Speed']
Y = df['Range']
plt.scatter(X,Y, color = 'green')
plt.xlabel('Bullet Speed')
plt.ylabel('Range')
plt.show()
```

```python
import statsmodels.api as sm
```

```python
results = sm.OLS(Y,X).fit()
print(results.summary())
#The results shows that the relationship is not strong. The value of coeffienct is 0.7336.
```

```python
sns.regplot(Y,X)
#The plot shows that there is a positively correlated relationship between a weapon's range
#and the bullet speed. My hypothesis is supported by this analysis. 
```
